```python
%matplotlib inline
from __future__ import division

import math
import matplotlib.pylab as plt
import numpy as np
import scipy.stats as stats
```

```python
# plot different probability distributions
def plot_1():
    x = np.linspace(-5, 5, num=75, endpoint=False)
    y = stats.norm.pdf(x)
    plt.plot(x, y, 'k')
    plt.savefig('pdf-normal.png')
    plt.show()

def plot_2():
    x = np.linspace(0, 1, num=75, endpoint=True)
    y = stats.beta.pdf(x, 2, 4)
    plt.plot(x, y, 'k')
    plt.savefig('pdf-beta.png')
    plt.show()

def plot_3():
    x = np.linspace(-5, 5, num=75, endpoint=False)
    y = stats.cauchy.pdf(x)
    plt.plot(x, y, 'k')
    x = np.linspace(-5, 5, num=75, endpoint=False)
    y = stats.norm.pdf(x)
    plt.plot(x, y, 'b--', alpha=0.2)
    plt.savefig('pdf-cauchy.png')
    plt.show()
    
plot_1()
plot_2()
plot_3()
```

```python
# calculate Fourier transform of square root of Cauchy distribution
def plot_4():
    n, m = 32, 16
    nm = n * m
    k1 = np.arange(0, nm)
    x = k1 / m - n / 2
    dist = stats.cauchy(loc=1, scale=1)
    f_half = np.array([math.sqrt(v) for v in dist.pdf(x)])
    dft = np.fft.fft(f_half)
    k = np.arange(0, nm)
    omega = -2 * np.pi / n * k
    phi = (1 / m) * np.exp(1j * np.pi * k) * dft

    # adjust data to plot range of interest
    phi_s = np.fft.fftshift(phi)
    ix0 = nm//2 - n
    ix1 = nm//2 + n
    plt.plot(omega[ix0:ix1] - omega[nm//2], np.real(phi_s[ix0:ix1]), 'b')
    plt.plot(omega[ix0:ix1] - omega[nm//2], np.imag(phi_s[ix0:ix1]), 'r')
    plt.savefig('phi.png')
    plt.show()
    return
plot_4()
```
